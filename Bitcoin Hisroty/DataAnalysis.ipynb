```python
def calculate_accuracy(y_true, y_pred, threshold=0.05):
    """
    Calculate the 'accuracy' for regression within a specified threshold.
    
    :param y_true: Actual values
    :param y_pred: Predicted values
    :param threshold: Threshold for considering a prediction 'accurate' (default is 5%)
    :return: Accuracy as a percentage
    """
    # Calculate the absolute percentage error
    percentage_error = np.abs((y_true - y_pred) / y_true)
    
    # Calculate 'accurate' predictions (within the threshold)
    accurate_predictions = percentage_error < threshold
    
    # Calculate accuracy
    accuracy = np.mean(accurate_predictions)
    
    return accuracy
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score
```

```python
scaler = MinMaxScaler()
# Select features to scale
features_to_scale = ['open', 'high', 'low', 'volume', 'SMA_12', 'SMA_26', 'RSI', 'close']

# Fit the scaler on the training data and transform both training and test data
data[features_to_scale] = scaler.fit_transform(data[features_to_scale])

# Split the data into features (X) and target (y)
X = data.drop('close', axis=1)
y = data['close']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Model Training (example with Random Forest)
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)

# Model Evaluation
predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
print(f"Mean Absolute Error: {mae}")

# Cross-validation
scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_absolute_error')
print(f"Cross-validated MAE: {-scores.mean()}")
```

```python
accuracy = calculate_accuracy(y_test, predictions, threshold=0.05)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
